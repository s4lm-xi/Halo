```python
import numpy as np
import pandas as pd
from datetime import datetime, timedelta
import os
import matplotlib.pyplot as plt
from PIL import Image
```

```python
def split_into_chunks(data, num_chunks=5):
    """
    Splits a list into approximately equal chunks.
    """
    data = list(data)
    chunk_size = len(data) // num_chunks
    remainder = len(data) % num_chunks
    
    chunks = []
    start = 0
    
    for i in range(num_chunks):
        end = start + chunk_size + (1 if i < remainder else 0)
        chunks.append(data[start:end])
        start = end
    
    return chunks

def calculate_chunk_differences(chunks):
    """
    Calculate the difference between the highest point and the next point to the right in each chunk.
    For the last chunk, calculate the difference between the highest and the last point.
    """
    differences = []
    
    for chunk in chunks:
        highest_point_index = np.argmax(chunk)
        if highest_point_index < len(chunk) - 1:
            difference = abs(chunk[highest_point_index] - chunk[highest_point_index + 1])
        else:
            difference = abs(chunk[highest_point_index] - chunk[-1])
        differences.append(difference)
    
    return differences

def find_earthquake_chunk(data, num_chunks=5):
    """
    Recursively split the data into chunks and identify the chunk where the earthquake likely occurred.
    """
    # Step 1: Split data into 5 chunks and calculate variance for each chunk
    chunks = split_into_chunks(data, num_chunks=num_chunks)
    noise_levels = [np.var(chunk) for chunk in chunks]

    # Step 2: Find the indices of the top 2 chunks with the highest variance (potential earthquake)
    top_two_indices = np.argsort(noise_levels)[-2:]
    
    # Step 3: For each of the top 2 chunks, split further and calculate differences
    smallest_difference = float('inf')
    earthquake_chunk = None
    earthquake_index_within_chunk = None
    earthquake_global_index = None
    
    start_idx = 0  # Track the starting index of each chunk in the original data

    for index in top_two_indices:
        print(index)
        # Track the starting index of the current chunk within the original data
        chunk_start_idx = start_idx
        
        # Step 4: Split the chunk into 5 smaller chunks
        refined_chunks = split_into_chunks(chunks[index], num_chunks=num_chunks)
        
        # Step 5: Calculate differences within each chunk
        differences = calculate_chunk_differences(refined_chunks)
        
        # Step 6: Find the chunk with the smallest difference
        min_difference = min(differences)
        if min_difference < smallest_difference:
            smallest_difference = min_difference
            earthquake_chunk = chunks[index]
            earthquake_index_within_chunk = np.argmax(earthquake_chunk)  # Find the highest point in the chunk
            earthquake_global_index = chunk_start_idx + earthquake_index_within_chunk  # Calculate global index

        # Update the starting index for the next chunk
        start_idx += len(chunks[index])
    
    return earthquake_chunk, smallest_difference, earthquake_global_index, index
```

```python
# Example Usage:
cat_directory = './data/lunar/training/catalogs/'
cat_file = cat_directory + 'apollo12_catalog_GradeA_final.csv'
cat = pd.read_csv(cat_file)
```

```python
index = 30
row = cat.iloc[index]
arrival_time = datetime.strptime(row['time_abs(%Y-%m-%dT%H:%M:%S.%f)'],'%Y-%m-%dT%H:%M:%S.%f')
test_filename = row.filename

data_directory = './data/lunar/training/data/S12_GradeA/'
csv_file = f'{data_directory}{test_filename}.csv'
data_cat = pd.read_csv(csv_file)

# Read in time steps and velocities
csv_times = np.array(data_cat['time_rel(sec)'].tolist())
csv_data = np.array(data_cat['velocity(m/s)'].tolist())

# Detect earthquake chunk
earthquake_chunk, smallest_diff, earthquake_global_index, detected_chunk_index = find_earthquake_chunk(csv_data)

# The global index corresponds to the position within the original data
earthquake_exact_time = csv_times[earthquake_global_index]

print("The exact time when the earthquake likely occurred:", earthquake_exact_time)
print("The global index of the earthquake in the original data:", earthquake_global_index)
print("Detected chunk index:", detected_chunk_index)

# Plot the full trace and the detected chunk on the same figure with two subplots
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 6))

# Plot the full trace on the first subplot
ax1.plot(csv_times, csv_data)
ax1.set_xlim([min(csv_times), max(csv_times)])
ax1.set_ylabel('Velocity (m/s)')
ax1.set_xlabel('Time (s)')
ax1.set_title(f'{test_filename}', fontweight='bold')

# Plot where the actual earthquake time is
arrival_line = ax1.axvline(x=row['time_rel(sec)'], c='green', label='Actual Earthquake Time')
ax1.legend(handles=[arrival_line])

# Plot the detected chunk on the second subplot
chunk_times = split_into_chunks(csv_times, num_chunks=5)[detected_chunk_index]
chunk_data = split_into_chunks(csv_data, num_chunks=5)[detected_chunk_index]

ax2.plot(csv_times, csv_data)
ax2.axvline(x=chunk_times[chunk_data.index(max(chunk_data))], c='red', linestyle='--', label='Predicted Earthquake Time')
ax2.set_xlim([min(csv_times), max(csv_times)])

ax2.set_ylabel('Velocity (m/s)')
ax2.set_xlabel('Time (s)')
ax2.set_title(f'Detected Chunk {detected_chunk_index}', fontweight='bold')
ax2.legend()
# Adjust layout to prevent overlap
plt.tight_layout()

# Show the combined plot
plt.savefig(f'{index}.png')
plt.show()
```
